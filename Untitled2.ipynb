```python
import jieba
```

```python
#先產生一個空字典準備裝詞彙
text_analy_dict = {}
```

```python
#將.txt內容用一個字串打包
with open('./speech.txt', encoding='utf-8') as speechfile:
    speech_string = speechfile.read()
```

```python
#用jieba.cut()方法將剛剛的字串分詞
for word in jieba.cut(speech_string):
    #如果分出來的詞在字典text_analy_dict裡，key為"詞"的值+1
    if word in text_analy_dict:
        text_analy_dict[word] += 1
    #如果分出來的詞不在字典text_analy_dict裡，增加以key為"詞"，值為1的物件
    else:
        text_analy_dict[word] = 1
```
